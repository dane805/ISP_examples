```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm
```

```python
mu, sigma, n = 70, 8, 5

norms = [norm.rvs(mu, sigma**2, n) for _ in range(50000)]
norms_normalized = [(np.mean(norm) - mu) / (np.std(norm, ddof=1)/(n**0.5)) for norm in norms]
norms_normalized = [norm for norm in norms_normalized if -3 < norm < 3]

xs = np.linspace(norm.ppf(0.001),
                norm.ppf(0.999), 1000)
ps_norm = norm.pdf(xs)
ps_t = t.pdf(xs, n-1)
```

```python
plt.hist(norms_normalized, bins=50, density=True, color='skyblue', alpha=0.8)
plt.plot(xs, ps_norm, color='red')
plt.plot(xs, ps_t, color='green')
plt.xlim(-3, 3)
plt.legend(['normal', 't'])
plt.savefig(f"./images/2.5.png")
```
