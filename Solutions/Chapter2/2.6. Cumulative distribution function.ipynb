```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson
```

```python
IMAGE_INDEX = 0

def norm_cdf_drawer(mu, sigma):
    '''
    x의 범위는 평균을 기준으로 3시그마 전후로 한다
    '''

    xs = np.arange(mu-3*sigma, mu+3*sigma, 0.1)
    ps_norm = [norm.cdf(x, mu, sigma) for x in xs]

    plt.plot(xs, ps_norm)
    plt.xlabel("x")
    plt.ylabel("P(X<x)")
    plt.title(f"Normal({mu}, {sigma}^2)")
    global IMAGE_INDEX
    plt.savefig(f"./images/2.6.{IMAGE_INDEX}.png")
    IMAGE_INDEX += 1

def poisson_cdf_drawer(lamb):
    '''
    x의 범위는 0부터 람다의 2배까지
    '''
    xs = range(lamb*2)
    ps_pois = [poisson.cdf(x, lamb) for x in xs]

    plt.plot(list(xs), ps_pois)
    plt.xlabel("x")
    plt.ylabel("P(X<x)")
    plt.title(f"Poisson({lamb})")
    global IMAGE_INDEX
    plt.savefig(f"./images/2.6.{IMAGE_INDEX}.png")
    IMAGE_INDEX += 1
```

```python
norm_cdf_drawer(7, 3)
```

```python
poisson_cdf_drawer(3)
```

```python
poisson_cdf_drawer(10)
```
